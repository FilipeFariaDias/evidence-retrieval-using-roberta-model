```python
weight_decay = 0.01
learning_rate = 3e-5
batch_size = 16
epochs = 1
num_classes = 2
l1_regularization = 0.1
gradient_accumulation_steps = 32
```

```python
class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self, name, fmt=':f'):
        self.name = name
        self.short_name = name.lower().replace(' ', '_')
        self.fmt = fmt
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count
```

```python
def add_metrics(dataset_name, bash_size, batch_index, iid_list, predicted_labels, true_labels, res, logits=None,
                order_list=None, trial_list=None, itype_list=None, genres_list=None, unlabeled=False,
                predicted_evidence=None, gold_evidence_label=None):
    """
    Add metrics to the res dictionary.

    Args:
        predicted_evidence:
        gold_evidence_label:
        genres_list:
        logits:
        dataset_name: str
        bash_size: int
        batch_index: int
        iid_list: list
        predicted_labels: tensor. shape: (bash_size, num_classes)
        true_labels: tensor.shape: (bash_size, )
        res: dict
        order_list: list
        trial_list: list
        itype_list: list
        unlabeled: bool

    Returns:
    """
    res["predicted_label"] += predicted_labels.argmax(1).cpu().numpy().tolist()
    if predicted_evidence is not None:
        res["predicted_evidence"] += predicted_evidence.argmax(1).cpu().numpy().tolist()

    if not unlabeled:
        res["gold_label"] += true_labels.cpu().numpy().tolist()
        if gold_evidence_label is not None:
            res["gold_evidence_label"] += gold_evidence_label.cpu().numpy().tolist()

    if logits is not None:
        res["logits"] += logits.cpu().numpy().tolist()

    offset = batch_index * bash_size
    res["iid"] += iid_list[offset:offset + bash_size]
    if dataset_name == "nli4ct":
        res["trial"] += trial_list[offset:offset + bash_size]
        res["order_"] += order_list[offset:offset + bash_size]
        res["itype"] += itype_list[offset:offset + bash_size]
    elif dataset_name == 'multinli':
        res["genre"] += genres_list[offset:offset + bash_size]
```

```python
def create_metrics_dict():
    return {"iid": [],
            "predicted_label": [],
            "trial": [],
            "itype": [],
            "genre": [],
            "order_": [],
            "gold_label": [],
            "predicted_evidence": [],
            "gold_evidence_label": [],
            "logits": []}.copy()
```

```python
def validate(dataloader, model, criterion, extra_info):
  losses = AverageMeter('Loss', ':1.5f')
  top = AverageMeter('Accuracy', ':2.3f')

  predictions = []
  labels = []

  iid_list, trial_list, order_list, genre_list, types_list = None, None, None, None, None
  if extra_info is not None:
      iid_list, types_list, trial_list, order_list, genre_list = extra_info

  res = create_metrics_dict()

  model.eval()
  with torch.no_grad():
    end = time.time()
    for idx, batch in enumerate(dataloader):
      bsz = batch[0].size(0)

      if torch.cuda.is_available():
        batch = tuple(t.cuda() for t in batch)

      inputs = {'input_ids': batch[0],
                'attention_mask': batch[1],
                'labels': batch[2]}

      output = model(**inputs)
      predicted_labels = output[1].view(output[1].size(), num_classes)
      true_labels = batch[2].view(-1)

      loss = criterion(predicted_labels, true_labels)

      if gradient_accumulation_steps > 1:
          loss /= gradient_accumulation_steps

      losses.update(loss.item(), bsz)
      acc = accuracy_score(true_labels.cpu().numpy(), predicted_labels.argmax(1).cpu().numpy())
      top.update(acc, bsz)

      end = time.time()

      predictions.extend(predicted_labels.cpu().numpy())
      labels.extend(true_labels.cpu().numpy())

  return predictions, labels, losses.avg, top.avg, res
```

```python
def train_roberta(dataloader, model, criterion, optimizer, scheduler, epoch, extra_info):
  predictions = []
  labels = []
  losses = AverageMeter('Loss', ':1.5f')
  top = AverageMeter('Accuracy', ':1.3f')


  l1_criterion = nn.L1Loss()

  iid_list, trial_list, order_list, genre_list, types_list = None, None, None, None, None
  if extra_info is not None:
      iid_list, types_list, trial_list, order_list, genre_list = extra_info

  res = create_metrics_dict()

  model.train()

  for idx, batch in enumerate(dataloader):
    bsz = batch[0].size(0)
    batch = tuple(t.cuda() for t in batch) if torch.cuda.is_available() else batch

    inputs = {'input_ids': batch[0],
              'attention_mask': batch[1],
              'labels': batch[2]}

    output = model(**inputs)
    predicted_labels = output[1].view(output[1].size(), num_classes)
    true_labels = inputs['labels'].view(-1)

    add_metrics(dataset_name='nli4ct', bash_size=bsz, batch_index=idx, iid_list=iid_list,
                predicted_labels=predicted_labels, true_labels=true_labels, res=res, logits=None,
                order_list=order_list, trial_list=trial_list, itype_list=types_list, genres_list=genre_list,
                unlabeled=False, predicted_evidence=None, gold_evidence_label=None)

    loss = criterion(predicted_labels, true_labels)

    if l1_regularization > 0:
      for param in model.parameters():
          loss += l1_regularization * l1_criterion(param, torch.zeros_like(param))

    if gradient_accumulation_steps > 1:
      loss /= gradient_accumulation_steps

    losses.update(loss.item(), bsz)
    loss.backward()

    acc = accuracy_score(true_labels.cpu().numpy(), predicted_labels.argmax(1).cpu().numpy())
    top.update(acc, bsz)

    if ((idx + 1) % gradient_accumulation_steps == 0) or ((idx + 1) == len(dataloader)):
      optimizer.step()
      scheduler.step((epoch + idx) / len(dataloader))
      optimizer.zero_grad()

    # predictions.extend(predicted_labels.cpu().numpy())
    # labels.extend(true_labels.cpu().numpy())

  return losses.avg, top.avg, res
```

```python
train_iids = [row.iid for _, row in train.iterrows()]
train_types = [row.itype for _, row in train.iterrows()] if 'itype' in train.columns else None
train_trials = [row.trial for _, row in train.iterrows()] if 'trial' in train.columns else None
train_orders = [row.order_ for _, row in train.iterrows()] if 'order_' in train.columns else None
train_genres = [row.genre for _, row in train.iterrows()] if 'genre' in train.columns else None

val_iids = [row.iid for _, row in val.iterrows()]
val_types = [row.itype for _, row in val.iterrows()] if 'itype' in val.columns else None
val_trials = [row.trial for _, row in val.iterrows()] if 'trial' in val.columns else None
val_orders = [row.order_ for _, row in val.iterrows()] if 'order_' in val.columns else None
val_genres = [row.genre for _, row in val.iterrows()] if 'genre' in val.columns else None
```

```python
import time
from sklearn.metrics import accuracy_score
from torch.nn import CrossEntropyLoss
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

best_val_acc = None
criterion = CrossEntropyLoss().cuda()
optimizer = AdamW(classifier.parameters(), lr=learning_rate,
                      weight_decay=weight_decay)

training_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
validation_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=5)

for epoch in range(epochs):

  time1 = time.time()

  train_loss, train_acc, train_results = train_roberta(training_loader, classifier, criterion, optimizer, scheduler, epoch, extra_info=(
      train_iids,
      train_types,
      train_trials,
      train_orders,
      train_genres
  ))

  time2 = time.time()

  print('Training epoch {}, total time {:.2f}, loss {:.7f}, acc {}'.format(epoch, (time2 - time1), train_loss, train_acc))

  # print(f'Training Precision: {precision_score(train_labels, train_predictions)}')
  # print(f'Training Recall: {recall_score(train_labels, train_predictions)}')
  # print(f'Training F1 Score: {f1_score(train_labels, train_predictions)}')


  val_time1 = time.time()

  val_predictions, val_labels, val_loss, val_acc, val_results = validate(validation_loader, classifier, criterion, extra_info=(
      val_iids,
      val_types,
      val_trials,
      val_orders,
      val_genres
  ))

  val_time2 = time.time()

  print('Validation epoch {}, total time {:.2f}, loss {:.7f}, acc {}'.format(epoch, (val_time2 - val_time1), val_loss, val_acc))

  # precision = precision_score(val_labels, val_predictions)
  # print(f'Validation Precision: {precision}')
  # print(f'Validation Recall: {recall_score(val_labels, val_predictions)}')
  # print(f'Validation F1 Score: {f1_score(val_labels, val_predictions)}')

  if best_val_acc is None or val_acc > best_val_acc:
    best_val_acc = val_acc
    print("New best accuracy: {:.3f}".format(best_val_acc.item()))
    save_file = os.path.join(MODELS_FOLDER_PATH, 'classifier_best.pth')
    state = {
        'models': classifier.state_dict(),
        'optimizer': optimizer.state_dict(),
        'best_metric': best_val_acc
    }
    torch.save(state, save_file)
    del state
  break

save_file = os.path.join(MODELS_FOLDER_PATH, 'classifier_last.pth')
state = {
    'models': classifier.state_dict(),
    'optimizer': optimizer.state_dict(),
    'best_metric': val_acc,
    'train_results': train_results,
    'val_results': val_results
}
torch.save(state, save_file)
#del state


```
